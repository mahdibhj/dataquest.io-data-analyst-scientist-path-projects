```python
dict ={}
unique_major = recent_grads['Major_category'].unique()
```

```python
print(unique_major)
```

```python
for maj in unique_major:
    major = recent_grads[recent_grads['Major_category']== maj]
    total = major['Total'].sum()
    
    dict[maj] = total
    
```

```python
sort_dict = sorted(dict.items(), key=lambda x: x[1], reverse=True)

for i in sort_dict:
	print(i[0], i[1])
```
